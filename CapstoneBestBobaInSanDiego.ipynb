## Best Boba in San Diego County

San Diego is home to 3,338,330 residents over an area of 4,526 sqmi. The regions we will focus on are North County, South County, East County, and Central San Diego.    

This information will be the most helpful for those in need of a boba drink with certain priorities in mind.

I want to prioritize reviews, parking, and accessibility.  

Reviews - Boba places that taste good
Parking - Places that have accessible parking 
Accessibility - Places that are easy to get to

#### Target Audience:
